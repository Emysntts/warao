```python
print("Listando dicts:")

modules_to_inspect = {
    "g1_dicionarios_separados": g1_dicionarios_separados,
    "g1_dicionarios_completos": g1_dicionarios_completos,
    "g2_dicionarios_separados": g2_dicionarios_separados,
    "g2_dicionarios_warao": g2_dicionarios_warao,
    "g3_dados": g3_dados,
}

all_available_dictionaries = {}
for mod_name, mod_obj in modules_to_inspect.items():
    print(f"\nDictionaries in {mod_name}:")
    found_dicts_in_mod = []
    for attr_name in dir(mod_obj):
        if not attr_name.startswith("__"):
            attr_value = getattr(mod_obj, attr_name)
            if isinstance(attr_value, dict):
                print(f"  - {attr_name} (Entries: {len(attr_value)})")
                found_dicts_in_mod.append(attr_name)
                all_available_dictionaries[f"{mod_name}.{attr_name}"] = attr_value
    if not found_dicts_in_mod:
        print("  (No dictionaries found in this module)")
```

```python
from utils import analyze_dictionary

print("\n" + "="*50)
print(" EDA PARA DICTS SELECIONADOS")
print("="*50 + "\n")

selected_dictionaries_to_analyze_names = [
    "g1_dicionarios_completos.dicionario_portugues",
    "g1_dicionarios_completos.dicionario_espanhol",
    "g2_dicionarios_warao.dicionario_portugues_para_warao",
    "g2_dicionarios_separados.dicionario_para_comecar_portugues",
    "g2_dicionarios_separados.dicionario_para_comecar_espanhol",
    "g3_dados.dicionario_numeros",
    "g3_dados.dicionario_frases_interacao",
    "g3_dados.dicionario_partes_corpo",
    "g3_dados.dicionario_saude",
    "g2_dicionarios_separados.dicionario_cores_portugues_warao"
]

for full_name in selected_dictionaries_to_analyze_names:
    module_name, dict_name = full_name.split('.', 1)
    if module_name in modules_to_inspect:
        dictionary_obj = getattr(modules_to_inspect[module_name], dict_name, None)
        analyze_dictionary(full_name, dictionary_obj)
    else:
        print(f"\n--- Module '{module_name}' for dictionary")
```

```python
from utils import unify_translations, count_syllables

translations = unify_translations(modules_to_inspect)

syllable_freq = count_syllables(translations)

print("Top 20 silabas mais comuns em Warao:")
for syl, freq in syllable_freq.most_common(20):
    print(f"  {syl}: {freq}")
```

```python
from collections import Counter
import matplotlib.pyplot as plt

chars = Counter(''.join(translations))
most_chars = chars.most_common(30)

print("Top 30 caracteres mais comuns em Warao:")
for ch, f in most_chars:
    print(f"  '{ch}': {f}")

plt.figure(figsize=(10,4))
plt.bar([c for c,_ in most_chars], [f for _,f in most_chars])
plt.title("Frequência de caracteres")
plt.show()
```

```python
import re

words = re.findall(r"\b[^\s]+\b", " ".join(translations))
lengths = [len(w) for w in words]

plt.figure(figsize=(8,4))
plt.hist(lengths, bins=range(1, max(lengths)+2), color='C1')
plt.xlabel("Tamanho da palavra (n caracteres)")
plt.ylabel("Frequência")
plt.title("Distribuição de tamanhos de palavras Warao")
plt.show()
```

```python
from itertools import islice

def char_ngrams(text, n):
    return (text[i:i+n] for i in range(len(text)-n+1))

# conta bigramas e trigramas
bigrams = Counter()
trigrams = Counter()
for w in words:
    bigrams.update(char_ngrams(w,2))
    trigrams.update(char_ngrams(w,3))

print("Top 10 bigramas:", bigrams.most_common(10))
print("Top 10 trigramas:", trigrams.most_common(10))
```

```python
from collections import defaultdict

rev = defaultdict(list)
for mod in modules_to_inspect.values():
    for attr in dir(mod):
        obj = getattr(mod, attr)
        if isinstance(obj, dict):
            for k,v in obj.items():
                if isinstance(v,str):
                    rev[v].append((mod.__name__, attr, k))

# filtra onde v aparece em >1 key
dups = {v:ks for v,ks in rev.items() if len(ks)>1}
print(f"Traduções repetidas: {len(dups)}")
for w, occurrences in list(dups.items())[:5]:
    print(w, "→", occurrences)
```
